## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
import pandas as pd
import numpy as np

customer_data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
customer_data
```

```python
new_customer_data_1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
new_customer_data_1
```

```python
new_customer_data_2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
new_customer_data_2
```

```python
#combine the three datasets and then apply the cleaning function to the new combined dataset

total_customer_data = pd.concat([customer_data, new_customer_data_1, new_customer_data_2], axis=0)
total_customer_data
```

```python
"""Column names should be in lower case"""

total_customer_data.columns = total_customer_data.columns.str.lower()
total_customer_data.columns
```

```python
"""White spaces in column names should be replaced by _"""
total_customer_data.columns = total_customer_data.columns.str.replace(' ', '_')
total_customer_data.columns
```

```python
"""Merge 'st' with 'state'"""
total_customer_data['state'] = total_customer_data['state'].fillna(total_customer_data['st'])

# Drop the redundant 'st' column
total_customer_data.drop(columns=['st'], inplace=True)
total_customer_data
```
