```python
import pandas as pd
import numpy as np

marketing_customer_analysis = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
marketing_customer_analysis.info()
```

```python
#1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000
#and have a response of "Yes" to the last marketing campaign.

# Pivot the DataFrame to see the customers based on the total_claim_amount>1000 and yes to marketing campaign

total_claim_amount_thresh = 1000

filtered_total_claim_amount = marketing_customer_analysis[marketing_customer_analysis['Total Claim Amount'] > total_claim_amount_thresh]
filtered_total_claim_amount

campaign_response = filtered_total_claim_amount[filtered_total_claim_amount['Response'] == 'Yes']
campaign_response 

#another way to do this:
#filtered_df = df[(df['Total Claim Amount']>1000) & (df['Response']=='Yes')]
```

```python
campaign_response = filtered_total_claim_amount[filtered_total_claim_amount['Response'] == 'Yes']
campaign_response 
```

```python
#Using the original Dataframe, 
#analyze the average total_claim_amount 
#by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. 
#Write your conclusions.

mean_tca_gen_yescamp = marketing_customer_analysis.pivot_table(values='Total Claim Amount', index='Policy Type', columns='Gender', aggfunc='mean')
mean_tca_gen_yescamp
```

```python
#Analyze the total number of customers who have policies in each state, 
#and then filter the results to only include states where there are more than 500 customers.

state_cust_count = marketing_customer_analysis.groupby(['State'])['Policy'].count()
state_cust_count
#policy_state_over_fivehun = marketing_customer_analysis[(sum(marketing_customer_analysis['State'] > 500)) and () total_claim_amount_thresh]
#filtered_total_claim_amount

#campaign_response = filtered_total_claim_amount[filtered_total_claim_amount['Response'] == 'Yes']
#campaign_response 
```
