```python
#treat the categorical data
categoricals = market_df.select_dtypes(include=object)
ordinals = categoricals[["coverage","education","vehicle_size"]]
nominals = categoricals.drop(columns=["coverage","education","vehicle_size"])

nominals
```

```python
enc = OrdinalEncoder(categories=[["Basic","Extended","Premium"]])

ordinals["coverage"] = enc.fit_transform(ordinals[["coverage"]])

enc = OrdinalEncoder(categories=[['High School or Below', 'College', 'Bachelor', 'Master', 'Doctor']])

ordinals["education"] = enc.fit_transform(ordinals[['education']])

enc = OrdinalEncoder(categories=[['Medsize', 'Small', 'Large']])

ordinals["vehicle_size"] = enc.fit_transform(ordinals[["vehicle_size"]])
```

```python
final_market_df = pd.concat([numericals_df,ordinals,nominals], axis=1)
final_market_df.info()
```

```python
#X-y split on the new concatenated data (y is the target variable, in this case, "total claim amount")

X = final_market_df.drop(columns=['total_claim_amount'])
y = final_market_df[['total_claim_amount']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 42)
```

```python
scaler = MinMaxScaler()
scaler_fit = scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#apply linear regression
lm = LinearRegression()
model = lm.fit(X_train_scaled, y_train)
```

```python
#Model Interpretation.
#find the slope of all the columns of the X_train dataframe
model.coef_
```

```python
#find the intercept
model.intercept_
```

```python
#predict total_claim_amount using the concated df
y_pred_final_market_df = model.predict(X_test)
y_pred_final_market_df = pd.DataFrame(y_pred_market_df)
y_pred_final_market_df
```

```python
y_test = y_test.reset_index(drop=True)
```

```python
#find the residuals
residuals_final_market_df = pd.concat([y_pred_final_market_df, y_test], axis = 1)
residuals_final_market_df = residuals_final_market_df.rename(columns = {"total_claim_amount": "y_test", 0: "y_pred"})
residuals_final_market_df["residuals"] = residuals_final_market_df["y_pred"] - residuals_final_market_df["y_test"]
residuals_final_market_df
```

```python
#root mean squared error 
#on average when we predict the total_claim_amount, we are this far off
np.sqrt(mse(y_test, y_pred_final_market_df))
```
